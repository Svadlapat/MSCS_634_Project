```python
import pandas as pd

# ✅ Correct working dataset URL (verified)
url = "https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv"

df = pd.read_csv(url)

print("✅ Dataset loaded successfully")
df.head()
```

```python
# ==============================
# Step 3: Data Cleaning
# ==============================

# Remove customerID (not useful for modeling)
df.drop("customerID", axis=1, inplace=True)

# Convert TotalCharges to numeric (fix datatype issue)
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors='coerce')

# Remove missing values
df.dropna(inplace=True)

# Check cleaned data
print("Shape after cleaning:", df.shape)
df.head()
```

```python
# ==============================
# Step 4: Encoding
# ==============================

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in df.select_dtypes(include='object').columns:
    df[col] = le.fit_transform(df[col])

df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Churn', data=df)
plt.title("Churn Distribution")
plt.show()
```

```python
plt.hist(df['tenure'], bins=30)
plt.title("Tenure Distribution")
plt.xlabel("Tenure")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.title("Monthly Charges vs Churn")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# ==============================
# Step 6: Train-Test Split
# ==============================

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = df.drop("Churn", axis=1)
y = df["Churn"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
# ==============================
# Step 7: Regression
# ==============================

from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score

model = Ridge()
model.fit(X_train, y_train)

print("Ridge R2 Score:", r2_score(y_test, model.predict(X_test)))
```

```python
# ==============================
# Step 8: Classification
# ==============================

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

clf = DecisionTreeClassifier(max_depth=5)
clf.fit(X_train, y_train)

print("Decision Tree Accuracy:", accuracy_score(y_test, clf.predict(X_test)))
```

```python
# ==============================
# Step 9: Clustering
# ==============================

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# Visualization
plt.scatter(df['tenure'], df['MonthlyCharges'], c=df['Cluster'])
plt.title("Customer Segmentation")
plt.xlabel("Tenure")
plt.ylabel("Monthly Charges")
plt.show()
```
